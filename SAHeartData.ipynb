```python
import pandas as pd
import numpy as np
```

```python
df_SAHeart = pd.read_csv("http://www-stat.stanford.edu/~tibs/ElemStatLearn/datasets/SAheart.data",
	sep=",")
```

```python
df_SAHeart.info()
```

```python
df_SAHeart.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
```

```python
sn.countplot(x='chd', hue='famhist', data=df_SAHeart)
```

Barplot showing the number of people having CHD or not depending on their family history of the disease

```python
sn.regplot(x='age', y='sbp', data = df_SAHeart)
```

The above graph shows that with the increase of age, systolic blood pressure also increases

```python
sn.distplot(df_SAHeart[df_SAHeart['chd']==0]['tobacco'], color='g', label='No Coronary heart disease')
sn.distplot(df_SAHeart[df_SAHeart['chd']==1]['tobacco'], color='r', label='With Coronary heart disease')
plt.legend()
```

The above graph shows the distribution of tobacco consumption for people with/without CHD.

```python
data_p=['sbp', 'obesity', 'age', 'ldl']
sn.pairplot(df_SAHeart[data_p], size =4)
```

From the above pairplot, we can conclude that obesity and age, obesity and ldl, ldl and sbp & obesity and sbp are correlated

```python
def age_name(age):
    if (age >= 0 and age <15):
        return ("Young")
    elif (age >=15 and age <35):
        return("Adult")
    elif (age>=35 and age <55):
        return ("Mild")
    else :
        return ("Old")
```

```python
df_SAHeart['AgeGroup'] = df_SAHeart['age'].apply(age_name)
```

```python
df_SAHeart.head()
```

Categorizing the data in age groups

```python
sn.countplot(x='chd', hue='AgeGroup', data=df_SAHeart, palette='Set3')
```

Showing CHD cases in different Age groups

```python
sn.boxplot(x='AgeGroup', y='ldl', data=df_SAHeart)
```

From the above comparison, we may conclude that there are outliers in all the age category, the median for Old group is higher than the Mild and young group.
